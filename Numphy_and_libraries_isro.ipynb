```python
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,0.2)
y=np.sin(x)
```

```python
x
y
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,label='linear')
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x,y_sin,label='Sine',linewidth=4.0)
plt.plot(x,y_cos,label='Cos')
plt.plot(x,y_cos,label='Cos_2')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
plt.close()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y_sin)
plt.subplot(1,2,2)
plt.plot(x,y_cos)
plt.show()
```
